# จงทำการเข้ารหัส
ตัวอย่างที่ 1 Encryption
ABC  <-- อักษรต้นฉบับที่รับเข้าเพื่อทำการเข้ารหัส
CDE <-- อักษรผ่านการเข้ารหัส

ตัวอย่างที่ 2 Decryption
ZAB  <-- อักษรต้นฉบับที่รับเข้าเพื่อทำการถอดรหัส
XYZ <-- อักษรผ่านการถอดรหัส

````python
import numpy as np
import math

#ถ้าให้ดีกว่านี้ให้ใช้วิธีแยกข้อความ เช่น ABC เป็น [A, B, C]

Decryption = []
  
# จำนวนสูงสุดที่รับ
n = 3  
print("ใส่อักษรเช่น A แล้ว enter แล้วพิมพ์ B แล้ว enter แล้วพิมพ์ C")
print("ใส่อักษรอังกฤษเท่านั้น")
# iterating till the range
for i in range(0, n):
    ToList = str(input().upper())  #รับข้อมูลและแปลงเป็นตัวใหญ่
    List = (ord(ToList))   #แปลงเป็น char (ascii)
    Decryption.append(List) #เก็บค่าใน Array
De = [2,2,2]
Decryption2 = np.array(Decryption)
De2 = np.array(De)
print(Decryption)
print("1.Encryption")
print("2.Decryption")
case = input("กรุณาเลือกเมนู (เลือก 1 หรือ 2): ")

if case == '1':
   sum = Decryption2 + De2  #ตั้งชื่อตัวแปรไม่ดี ขอโทษ  "Decryption2"
   print("คุณเลือก Encryption")
elif case == '2':
   sum = Decryption2 - De2
   print("คุณเลือก Decryption")
else:
   print("กรุณาใส่เลข 1 - 2")

A = sum[0]
B = sum[1] 
C = sum[2] 
if A < 65 or A > 90:
  if A < 65:
    A = A +26
  elif A > 90:
    A = A -26
if B < 65 or B > 90:
  if B < 65:
    B = B +26
  elif B > 90:
    B = B -26
if C < 65 or C > 90:
  if C < 65:
    C = C +26
  elif C > 90:
    C = C -26

# print(A)
# print(B)
# print(C)
#จะให้เป็น Array หรือ loop ก็ได้ จะสวยกว่า print ดิบๆแบบนี้
print(chr(A),chr(B),chr(C))
# print(chr(A))
# print(chr(B))
# print(chr(C))
````



> จงเขียนโปรแกรมแสดงผลจำนวนเฉพาะที่มีตั้งแต่ 1 ถึง 1,000



````python
def prime_number(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

start = 1
end = 1000

primes = []
for i in range(start, end + 1):
    if prime_number(i):
        primes.append(i)
print(primes)
````



> จงสร้างโปรแกรมประมวลผลแฟ้มข้อมูลให้สามารถ บันทึกข้อมูล(Write Data)  อ่านข้อมูล(Read Data)    เพิ่มข้อมูล(Append Data)   โดยทั้ง 3 กิจกรรมต้องอยู่ในไฟล์เดียวกัน



````python
print("โปรแกรมเพิ่มลบแก้ไขไฟล์")


#แนะนำให้รันในเครื่องตัวเองโค้ดนี้

def switch_case(case):
    if case == '1':
        print("ทำการสร้างไฟล์ กรุณาระบุข้อความในไฟล์ (ภาษาอังกฤษเท่านั้น)")
        insert = (input("ใส่ข้อความที่เพิ่ม: "))
        with open('test.txt', 'w') as f:  #สร้างไฟล์ test.txt ขึ้นมา
            f.writelines([insert]) #ฟังค์ชั่นสร้างไฟล์
            f.close() #ปิดการทำงาน
    elif case == '2':
        with open('test.txt', 'r') as f:
            line = f.readline()
            while line:
                print(line)
                line = f.readline() #ฟังค์ชั่นอ่านข้อมูล
                f.close()
    elif case == '3':
        print("เพิ่มข้อมูล")
        print("กรุณาระบุใส่ข้อความที่เพิ่ม (ภาษาอังกฤษเท่านั้น)")
        update = (input("ข้อความ: "))
        with open('test.txt', 'a') as f:
            f.writelines([update]) #ฟังค์ชั่นเพิ่มข้อมูล
            f.close() #ปิดการทำงาน
    else:
        print("กรุณาใส่เลข 1 - 3")

print("1. สร้างไฟล์\n2. อ่านไฟล์\n3. เพิ่มข้อมูล")
print("พิมพ์เลข 1 2 3 เพื่อเลือกเมนู")
user_select = (input("ระบุ: "))
switch_case(user_select)
````
